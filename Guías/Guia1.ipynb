# **2. Operadores y Condicionales**

Ejercicio 1:
Escribir un programa que tome una edad como entrada y verifique si una persona es
menor de edad, mayor de edad o tiene más de 65 años. Mostrar un mensaje
adecuado para cada caso.

```python
edad = int(input("Ingrese una edad: "))
if edad < 18:
  print("La persona es menor de edad")
elif edad > 18 and edad <= 65:
  print("La persona es mayor de edad")
else:
  print("La persona es mayor de 65 años")
```

Ejercicio 2:
Crea un programa que tome dos números como entrada y verifique cuál es mayor,
luego informarlo con el número. Si son iguales, mostrar un mensaje que indique que
son iguales.

```python
num1 = int(input("Ingrese un número entero: "))
num2 = int(input("Ingrese un número entero: "))
if num1 == num2:
  print("Los números ingresados son  iguales")
elif num1 > num2:
  print(f"{num1} es mayor a {num2}")
else:
  print(f"{num2} es mayor a {num1}")
```

Ejercicio 5:
Crear un programa que almacene el diccionario con los créditos de las asignaturas
de un curso {'Matemáticas': 6, 'Física': 4, 'Química': 5} y después muestre por pantalla
los créditos de cada asignatura en el formato *asignatura* tiene *créditos* créditos,
donde *asignatura* es cada una de las asignaturas del curso, y *créditos* son sus
créditos. Al final debe mostrar también el número total de créditos del curso.

```python
curso = {
    "Matemáticas": 6,
    "Física": 4,
    "Química": 5
}

total = 0;

for asignatura in curso:
  print(f"{asignatura} tiene {curso.get(asignatura)} créditos")
  total+= curso.get(asignatura)
print()
print(f"El total de créditos del curso es: {total}")
```

Ejercicio 6:
Crear un programa que pida al usuario una letra indique si es vocal o no.

```python
vocales = ("a", "e", "i", "o", "u")
char = input("Ingrese una letra: ")
for vocal in vocales:
  if char == vocal:
    print(f"La letra ingresada es la vocal '{vocal}'")
    break
else:
  print("La letra ingresada no es una vocal")
```

Ejercicio 7:
Crear un programa en el que se pregunte al usuario por una frase y una letra, y
muestre por pantalla el número de veces que aparece la letra en la frase.

```python
cant = 0
frase = input("Por favor ingrese una frase: ")
letra = input("Por favor ingrese una letra: ")
for char in frase:
  if char == letra:
    cant+=1
print(f"La letra ingresada ({letra}) aparece {cant} veces en la frase que se ingresó.")
```
